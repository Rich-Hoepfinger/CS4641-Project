```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("zscore_normalised_trackdata.csv")
df = df.iloc[:,4:].drop('id', axis = 1)
df.head()
```

```python
features = df.iloc[:,:-1]
corrvals = features.corr()
fig, ax = plt.subplots(figsize=(10,8))
# Collinearity Matrix
sn.heatmap(corrvals, annot = True, ax = ax)
```

```python
features.drop(["energy", "loudness"], axis = 1, inplace = True)
```
